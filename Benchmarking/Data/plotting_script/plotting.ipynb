Fixed depth experiments: 

```python
plt.figure(figsize=(8,5))

FIXED_DEPTH = 4
dm_purity_values = []
model_purity_values = []
experiments = ['Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10', 'Q11', 'Q12', 'Q13', 'Q14', 'Q15', 'Q16']
for experiment in experiments:
    with open(f'../../Data_test/DensityMatrices_metrics/{experiment}.json') as f:
        data1 = json.load(f)

    with open(f'../../Data_test/AnalyticalModel_metrics/{experiment}_purity_model_globalDP_p1_p2.json') as f:
        data4 = json.load(f)
    
    x = np.arange(len(data1['all_pur_diff_n']))  # depth axis for DM, CS & ST

    x4 = data4['depth_tab_more_points'] # global DP
    
    # Extract the purity value at the fixed depth
    dm_purity_values.append(data1['all_pur_diff_n'][FIXED_DEPTH])
    
    # Find indices where depth is in the range [FIXED_DEPTH, FIXED_DEPTH + 1)
    filtered_indices = [i for i, d in enumerate(x4) if FIXED_DEPTH <= d < FIXED_DEPTH +1]
    
    if filtered_indices:
        # Get the purity values for this depth range (average them or take first one)
        filtered_purity_values = [data4['all_pur_diff_n'][i] for i in filtered_indices]
        # Taking the average of purity values in the depth range
        model_purity_values.append(np.mean(filtered_purity_values))
    else:
        # If no data found in the depth range
        model_purity_values.append(None)
    
x = np.arange(2, 17)  # x-axis shows qubit numbers from 2 to 16

plt.plot(x, dm_purity_values, 'k-o', label=f'Exact (Depth {FIXED_DEPTH})', linewidth=2)
plt.plot(x, model_purity_values, 'r--s', label=f'Purity Model - global_DP (Depth {FIXED_DEPTH})', linewidth=2)

plt.xlabel('Qubits')
plt.ylabel('Purity')
plt.title(f'Purity at Depth {FIXED_DEPTH}')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(x)
plt.show()
```
