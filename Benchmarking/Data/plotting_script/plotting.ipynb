```python
import json
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
matplotlib.use('TkAgg') 
```

Single experiment set up

```python
# Make sure you're inside the plotting_script directory

# Pick the experiments to plot
with open('../../Data_test/DensityMatrices_metrics/Q3.json') as f:
    data1 = json.load(f)

with open('../../Data_test/Swap_metrics/Q3_m3000_g3_s3.json') as f:
    data2 = json.load(f)

with open('../../Data_test/ClassicalShadows_metrics/Q3_m320_k1000_g5_s3.json') as f:
    data3 = json.load(f)

with open('../../Data_test/AnalyticalModel_metrics/Q3_purity_model_globalDP.json') as f:
    data4 = json.load(f)

with open('../../Data_test/AnalyticalModel_metrics/Q3_purity_model_globalDP_CS.json') as f:
    data5 = json.load(f)

```

```python
x = np.arange(len(data1['all_pur_diff_n']))  # depth axis for DM, CS & ST

x4 = data4['depth_tab_more_points'] # global DP
x5 = data5['depth_tab_more_points'] # global DP + CS
```

Plot purity

```python
plt.figure(figsize=(8,5))
plt.plot(x, data1['all_pur_diff_n'], 'k-', label='exact')
# plt.errorbar(x, data2['all_pur_mean_diff_n'], yerr=data2['all_pur_std_diff_n'],
#               fmt='none',ecolor='r', capsize=5, label='Swap')
plt.errorbar(x, data3['all_pur_mean_diff_n'], yerr=data3['all_pur_std_diff_n'],
             fmt='none', ecolor='b', capsize=5, label='Classical Shadows') 
#plt.plot(x4, data4['all_pur_diff_n'], label='Purity Model - global_DP', linestyle='dotted')
plt.plot(x5, data5['all_pur_diff_n'], label='Purity Model - global_DP_CS', linestyle='dotted')

plt.xlabel('Depth')
plt.ylabel('Purity')
plt.title('Purity')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(x)
plt.show()
```

Plot Renyi-2 entropy

```python
plt.figure(figsize=(8,5))
plt.plot(x, data1['all_R2d_diff_n'], 'k-', label='exact')
# plt.errorbar(x, data2['all_R2d_mean_diff_n'], yerr=data2['all_R2d_std_diff_n'],
#               fmt='none',ecolor='r', capsize=5, label='Swap')
plt.errorbar(x, data3['all_R2d_mean_diff_n'], yerr=data3['all_R2d_std_diff_n'],
             fmt='none', ecolor='b', capsize=5, label='Classical Shadows') 
# plt.plot(x4, data4['all_R2d_diff_n'], label='Purity Model - global_DP', linestyle='dotted')
plt.plot(x5, data5['all_R2d_diff_n'], label='Purity Model - global_DP_CS', linestyle='dotted')

plt.xlabel('Depth')
plt.ylabel('Renyi-2')
plt.title('Renyi-2 entropy')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(x)
plt.show()
```

Multiple experiments set up

```python
plt.figure(figsize=(8,5))

for experiment in ['Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7']:
    with open(f'../../Data_test/DensityMatrices_metrics/{experiment}.json') as f:
        data1 = json.load(f)

    with open(f'../../Data_test/AnalyticalModel_metrics/{experiment}_purity_model_globalDP.json') as f:
        data4 = json.load(f)
    
    x = np.arange(len(data1['all_pur_diff_n']))  # depth axis for DM, CS & ST

    x4 = data4['depth_tab_more_points'] # global DP
    
    plt.plot(x, data1['all_pur_diff_n'], 'k-', label='exact')
    plt.plot(x4, data4['all_pur_diff_n'], label='Purity Model - global_DP - {}'.format(experiment), linestyle='dotted')

plt.xlabel('Depth')
plt.ylabel('Purity')
plt.title('Purity')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.xticks(x)
plt.show()
```
